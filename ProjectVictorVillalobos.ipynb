<h4 style="color:#6e9a81; font-family: Times New Roman"> ¿Cuál es el ingreso anual de las propiedades vendidas?</h4>

<h4 style="color:#6e9a81; font-family: Times New Roman"> <h4 style="color:#6e9a81; font-family: Times New Roman"> ¿Cuál es el costo promedio de las propiedades residenciales y comerciales?</h4>

```python
# Calcular el costo promedio de las propiedades residenciales
promedio_residential= round(dfDataP[dfDataP['property type'] == 'Residential']['sale amount'].mean())

# Calcular el costo promedio de las propiedades comerciales
promedio_commercial = round(dfDataP[dfDataP['property type'] == 'Commercial']['sale amount'].mean())

print(f"Costo promedio de las propiedades residenciales: {promedio_residential}")
print(f"Costo promedio de las propiedades comerciales: {promedio_commercial}")
```

<h4 style="color:#6e9a81; font-family: Times New Roman"> <h4 style="color:#6e9a81; font-family: Times New Roman"> ¿Cuál es tipo de propiedades residenciales que se vende con mayor frecuencia, donde y con que costo promedio?</h4>

```python
# Identificar el tipo de propiedad residencial que se vende con mayor frecuencia
tipo_frecuente = dfDataP[dfDataP['property type'] == 'Residential']['residential type'].mode()[0]

# Identificar la ubicación donde se vende el tipo de propiedad residencial más frecuente
ubicacion_frecuente = dfDataP[dfDataP['residential type'] == tipo_frecuente]['town'].mode()[0]

# Calcular el costo promedio del tipo de propiedad residencial más frecuente
costo_promedio = round(dfDataP[(dfDataP['property type'] == 'Residential') & (dfDataP['residential type'] ==tipo_frecuente)]['sale amount'].mean())

# Imprimir los resultados
print(f"Tipo de propiedad residencial que se vende con mayor frecuencia: {tipo_frecuente}")
print(f"Ubicación donde se vende el tipo de propiedad residencial más frecuente: {ubicacion_frecuente}")
print(f"Costo promedio del tipo de propiedad residencial más frecuente: {costo_promedio}")
```

<h4 style="color:#6e9a81; font-family: Times New Roman"> <h4 style="color:#6e9a81; font-family: Times New Roman"> ¿Cuál es margen promedio de ganancia por cada una de las propiedades vendidas?</h4>

```python
# Calcular el resultado de cada tipo de propiedad
def calculate_result(dfDataP, property_type):
# Filtrar las propiedades por tipo
    tipo_Pfiltrado = dfDataP[dfDataP['property type'] == property_type]
#Sumar los valores de cada columna
    total_assessed_value = tipo_Pfiltrado['assessed value'].sum()
    total_sale_amount = tipo_Pfiltrado['sale amount'].sum()
# Restar los totales de las dos columnas
    resultado_resta = total_sale_amount - total_assessed_value
# Dividir el resultado entre el número total de filas
    total_filas = tipo_Pfiltrado.shape[0]
    resultado_final = resultado_resta / total_filas
    return resultado_final
# Calcular el resultado para cada tipo de propiedad
property_types = dfDataP['property type'].unique()
resultados = {property_type: calculate_result(dfDataP, property_type) for property_type in property_types}

# Imprimir los resultados en una lista hacia abajo, sin paréntesis y eliminando el 'np.float64'
for property_type, resultado in resultados.items():
    print(f"{property_type}: {resultado:.2f}")
```

```python
# Generar un gráfico con los resultados
plt.figure(figsize=(12, 8))
plt.bar(resultados.keys(), resultados.values(), color='#e1918e')
plt.xlabel('Tipo de Propiedad')
plt.ylabel('Margen de Ganancia')
plt.title('Ganancia por tipo de Propiedad')
plt.xticks(rotation=45)
plt.yticks(range(0, int(max(resultados.values())) + 1, int(max(resultados.values()) / 10)))
plt.tight_layout()

# Ajustar el tamaño de las barras para que no se vean tan pequeñas
for i in range(len(resultados)):
    plt.text(i, list(resultados.values())[i], f"{list(resultados.values())[i]:.2f}", ha='center', va='bottom')

# Mostrar el gráfico
plt.show()
```
